```python
def bisec(func , x_max : float, x_min : float, LIMIT = 1e-20 ):
    
    x_mid = ( x_max + x_min ) / 2  
    
    while( np.abs( x_mid - x_max ) > LIMIT ):
        
        if func( x_mid ) > 0:
            x_max = x_mid
            
        elif func( x_mid ) == 0 :
            break
        else:
            x_min = x_mid
        print(x_mid)
        x_mid =( x_min + x_max ) / 2
        
    print(x_min)
```

```python
bisec(func, 1.5, 1.2)
```
